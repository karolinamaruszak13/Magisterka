```python
df_tweets = pd.read_sql("select place_id, tweet_text from tweets WHERE like_count > 200", mydb)
df_places = pd.read_sql("select id, name,country, country_code from places", mydb)
```

```python
sentiment_map = pd.merge(left=df_tweets, right=df_places, left_on='place_id', right_on='id')
sentiment_map
```

```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import nltk
nltk.downloader.download('vader_lexicon')
```

```python
sentiment_map['compound'] = sentiment_map['tweet_text'].apply(lambda x:SentimentIntensityAnalyzer().polarity_scores(x)['compound'])
sentiment_map['label'] = 0
for ind in sentiment_map.index:
    if sentiment_map['compound'][ind] < -0.05:
        sentiment_map['label'][ind] = 'negative'
    elif sentiment_map['compound'][ind] > 0.05:
        sentiment_map['label'][ind] = 'positive'
    else:
        sentiment_map['label'][ind] = 'neutral'
```

```python
sentiment_map['cityANDcountry'] = sentiment_map['name'] +','+ sentiment_map['country']
sentiment_map.drop_duplicates(subset ="id", keep = False, inplace = True)
sentiment_map
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_user_agent")
sentiment_map['city_coord'] = sentiment_map['cityANDcountry'].apply(geolocator.geocode)
sentiment_map
```

```python
sentiment_map = sentiment_map.dropna()
sentiment_map
```

```python
sentiment_map['latitude'] = sentiment_map['city_coord'].apply(lambda x: x.latitude)
sentiment_map['longitude'] = sentiment_map['city_coord'].apply(lambda x: x.longitude)
sentiment_map
```

```python
import mapclassify
import geoplot


geoplot.choropleth(sentiment_map,
                   hue="Perceptions of corruption",
                   scheme=mapclassify.Quantiles(sentiment_map["label"], k=3),
                   cmap="RdYlGn", edgecolor="white",
                   legend_labels=["negative", "neutral", "positive"],
                   legend_kwargs={"loc":"best",
                                  "fontsize": "large",
                                  "title":"Perception of Corruption",
                                  "title_fontsize":"large"},
                   figsize=(15,7), legend=True,
                  )

plt.title("Geograficzna analiza sentymentu", fontdict={"fontsize": 15}, pad=5)
```

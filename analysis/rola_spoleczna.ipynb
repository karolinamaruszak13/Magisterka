```python
df_users = pd.read_sql("select id, description, followers_count from users", mydb)
df_tweets = pd.read_sql("select author_id, tweet_text from tweets WHERE like_count > 200", mydb)
```

```python
username_tweet = pd.merge(left=df_tweets, right=df_users, left_on='author_id', right_on='id')
```

```python
username_tweet['description'] = username_tweet['description'].str.lower() 
username_tweet
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
import nltk
nltk.downloader.download('vader_lexicon')
```

```python
username_tweet['compound'] = username_tweet['tweet_text'].apply(lambda x:SentimentIntensityAnalyzer().polarity_scores(x)['compound'])
username_tweet['label'] = 0
for ind in username_tweet.index:
    if username_tweet['compound'][ind] < -0.05:
        username_tweet['label'][ind] = 'negative'
    elif username_tweet['compound'][ind] > 0.05:
        username_tweet['label'][ind] = 'positive'
    else:
        username_tweet['label'][ind] = 'neutral'
```

```python
username_tweet
```

```python
roles = []
for desc in username_tweet['description']:
    if "student" in desc:
        roles.append("student")
    elif "musician" in desc:
        roles.append("musician")
    elif "doctor" in desc:
        roles.append("doctor")
    elif "actress" in desc:
        roles.append("actress")
    elif "model" in desc:
        roles.append("model")
    elif "professor" in desc:
        roles.append("professor")
    elif "writer" in desc:
        roles.append(" writer ")
    elif "analyst" in desc:
        roles.append("analyst")
    elif "lawyer" in desc:
        roles.append("lawyer")
    elif "software developer" in desc:
        roles.append("software developer")
    elif "manager" in desc:
        roles.append("manager")
    elif "physican" in desc:
        roles.append("physican")
    elif "artist" in desc:
        roles.append("artist")
    elif "engineer" in desc:
        roles.append("engineer")
    elif "scientist" in desc:
        roles.append("scientist")
    elif "teacher" in desc:
        roles.append("teacher")
    elif "politician" in desc:
        roles.append("politician")
    else:
        roles.append("None")

username_tweet['role'] =  roles    
username_tweet = username_tweet[username_tweet.role != "None"]
username_tweet
```

```python
number_of_roles = pd.DataFrame(username_tweet.groupby('role').size())
number_of_roles
```

```python
import matplotlib.pyplot as plt
actress = int(number_of_roles.values[0])
analyst = int(number_of_roles.values[1])
artist = int(number_of_roles.values[2])
doctor = int(number_of_roles.values[3])
engineer = int(number_of_roles.values[4])
lawyer = int(number_of_roles.values[5])
manager = int(number_of_roles.values[6])
model = int(number_of_roles.values[7])
musician = int(number_of_roles.values[8])
professor = int(number_of_roles.values[9])
scientist = int(number_of_roles.values[10])
student = int(number_of_roles.values[11])
writer = int(number_of_roles.values[12])
role_names = ['actress', 'analyst', 'artist', 'doctor', 'engineer', 'lawyer', 'manager', 'model', 'musician', 'professor', 'scientist', 'student', 'writer']
roles_count = [actress, analyst, artist, doctor, engineer, lawyer, manager, model, musician, professor, scientist, student, writer]

plt.figure(figsize=(15,10))
plt.rcParams['font.size'] = '14' 
rects = plt.bar(role_names, roles_count, width=0.6, color='mediumblue')
plt.bar_label(rects)
plt.xticks(rotation = 90)
plt.ylabel('Ilość osób o określonej roli społecznej')
plt.title('Analiza ilości konkretnych ról społecznych oznaczonych przez użytkowników w opisie profilu')
plt.savefig(f'charts/Ilosc_rol_spolecznych.png')
plt.close()


```

```python
sentiment_by_role = username_tweet.groupby(['label', 'role']).size()
```

```python
import numpy as np
x = sentiment_by_role['negative'].index
x_axis = np.arange(len(x))


negative = sentiment_by_role['negative'].values
positive = sentiment_by_role['positive'].values
neutral = sentiment_by_role['neutral'].values

plt.figure(figsize=(15,10))
plt.rcParams['font.size'] = '14'
plt.bar(x_axis-0.2, negative, width=0.2, color='r', align='center', label='negatywny')
plt.bar(x_axis, positive, width=0.2, color='g', align='center', label='pozytywny')
plt.bar(x_axis+0.2, neutral, width=0.2, color='yellow', align='center', label='neutralny')
plt.xticks(rotation = 90)

plt.ylabel('Ilość tweetów o określonym sentymencie')
plt.title('Analiza sentymentu tweetów napisanych przez osoby z różnych grup społecznych')

plt.xticks(x_axis, x)
plt.legend()
plt.savefig(f'charts/sentymen_rol_spolczenych.png')
plt.close()
```

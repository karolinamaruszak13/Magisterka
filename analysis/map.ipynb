```python
import mysql.connector
import json
import pandas as pd
import datetime

mydb = mysql.connector.connect(

  host="localhost",
  user="root",
  password="Kaja25802580",
  database="twitter_database"
)

mycursor = mydb.cursor()
```

```python
df_placeCoordinates = pd.read_sql("select id from place_coordinates", mydb)
df_places = pd.read_sql("select id, name,country, country_code from places", mydb)
```

```python
merged_df= pd.merge(left=df_placeCoordinates, right=df_places, left_on='id', right_on='id')
merged_df['cityANDcountry'] = merged_df['name'] +','+ merged_df['country']
merged_df.drop_duplicates(subset ="id", keep = False, inplace = True)
merged_df
```

```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="my_user_agent")
merged_df['city_coord'] = merged_df['cityANDcountry'].apply(geolocator.geocode)
merged_df
```

```python
merged_df = merged_df.dropna()
merged_df
```

```python
merged_df['latitude'] = merged_df['city_coord'].apply(lambda x: x.latitude)
merged_df['longitude'] = merged_df['city_coord'].apply(lambda x: x.longitude)
merged_df
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm
from mpl_toolkits.basemap import Basemap
from matplotlib.patches import Polygon
from matplotlib.collections import PatchCollection
from matplotlib.colors import Normalize
```

```python
fig, ax = plt.subplots(figsize=(23,23))
coordinate_df = merged_df
m=Basemap(llcrnrlon=-180, llcrnrlat=-65, urcrnrlon=180, urcrnrlat=80)
m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
m.fillcontinents(color='grey', alpha=0.3)
m.drawcoastlines(linewidth=0.1, color="white")

# prepare a color for each point depending on the continent.
coordinate_df['labels_enc'] = pd.factorize(coordinate_df['country'])[0]
 
# Add a point per position
m.scatter(
    x=coordinate_df['longitude'], 
    y=coordinate_df['latitude'], 
    alpha=0.4, 
    c=coordinate_df['labels_enc'], 
    cmap="Set1"
)
plt.text( -175, -62,'Lokalizacja najpopularniejszych tweetów wykonana w Pythonie z użyciem Basemap', ha='left', va='bottom', size=9, color='#555555' );
plt.savefig('charts/loc_map.png', dpi=200)
plt.close()
```
